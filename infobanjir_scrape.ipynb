```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up WebDriver (Chrome in this case)
options = Options()
options.headless = True  # Run in headless mode (without a browser window)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Navigate to the website
url = 'https://publicinfobanjir.water.gov.my/aras-air/data-paras-air/?state=JHR&lang=en'
driver.get(url)

# Wait for the table to load (use explicit wait)
wait = WebDriverWait(driver, 10)

# Wait for any row of the table to be present
wait.until(EC.presence_of_element_located((By.XPATH, "//table[@id='normaltable1']//tbody//tr")))

# Locate the table rows
rows = driver.find_elements(By.XPATH, "//table[@id='normaltable1']//tbody//tr")

# Extract data for 'wl' column from each row
data = []
for row in rows:
    try:
        # Extract the 'wl' value (the number inside the <a> tag)
        wl = row.find_element(By.XPATH, ".//td[@data-th='wl']/a").text  # Extract the number inside <a>
        wl_link = row.find_element(By.XPATH, ".//td[@data-th='wl']/a").get_attribute('href')  # Get the link

        # Add the extracted data to the list
        data.append({
            'WL Value': wl,
            'WL Link': wl_link  # Link to the graph
        })

    except Exception as e:
        print(f"Error extracting data for row: {e}")

# Print the extracted data
for entry in data:
    print(entry)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up WebDriver (Chrome in this case)
options = Options()
options.headless = True  # Run in headless mode (without a browser window)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Navigate to the website
url = 'https://publicinfobanjir.water.gov.my/aras-air/data-paras-air/?state=JHR&lang=en'
driver.get(url)

# Wait for the table to load (use explicit wait)
wait = WebDriverWait(driver, 10)

# Wait for any row of the table to be present
wait.until(EC.presence_of_element_located((By.XPATH, "//table[@id='normaltable1']//tbody//tr")))

# Locate the table rows
rows = driver.find_elements(By.XPATH, "//table[@id='normaltable1']//tbody//tr")

# Extract combined data for each row
combined_data = []
for row in rows:
    try:
        # Extract data from each column
        no = row.find_element(By.XPATH, ".//td[@data-th='No']").text
        station_id = row.find_element(By.XPATH, ".//td[@data-th='Station ID']").text
        station_name = row.find_element(By.XPATH, ".//td[@data-th='Station Name']").text
        district = row.find_element(By.XPATH, ".//td[@data-th='District']").text
        main_basin = row.find_element(By.XPATH, ".//td[@data-th='Main Basin (mm)']").text
        sub_river_basin = row.find_element(By.XPATH, ".//td[@data-th='Sub River Basin (mm)']").text
        last_updated = row.find_element(By.XPATH, ".//td[@data-th='Last Update']").text
        normal = row.find_element(By.XPATH, ".//td[@data-th='Normal']").text
        alert = row.find_element(By.XPATH, ".//td[@data-th='Alert']").text
        warning = row.find_element(By.XPATH, ".//td[@data-th='Warning']").text
        danger = row.find_element(By.XPATH, ".//td[@data-th='Danger']").text

        wl = row.find_element(By.XPATH, ".//td[@data-th='wl']/a").text  # Extract the 'wl' value
        wl_link = row.find_element(By.XPATH, ".//td[@data-th='wl']/a").get_attribute('href')  # Get the link

        # Combine all data into a dictionary
        combined_data.append({
            'no': no,
            'station_id': station_id,
            'station_name': station_name,
            'district': district,
            'main_basin': main_basin,
            'sub_river_basin': sub_river_basin,
            'last_updated': last_updated,
            'wl': wl,
            'normal': normal,
            'alert': alert,
            'warning': warning,
            'danger': danger,
            'wl_link': wl_link
            
        })

    except Exception as e:
        print(f"Error extracting data for row: {e}")

# Print the combined data
for entry in combined_data:
    print(entry)

# Close the browser
driver.quit()
```
