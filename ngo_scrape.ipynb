```python
import requests
from bs4 import BeautifulSoup

def scrape_ngo_data():
    ngo_data_list = []

    # Loop through pages 1 to 8
    for page in range(1, 9):  # Pages 1 to 8
        print(f"Scraping page {page}...")

        # Making a GET request for the current page
        url = f'https://ngobase.org/st/MY.JO/johor-ngos-charities?page={page}'
        r = requests.get(url)

        # Check status code for response received
        if r.status_code != 200:
            print(f"Request failed with status code {r.status_code}")
            continue  # Skip to the next page if the request fails

        # Parsing the HTML
        soup = BeautifulSoup(r.content, 'html.parser')

        # Extracting all NGO entries
        ngo_listings = soup.find_all('div', class_='my_border ngo_listing_div')

        # Looping through each NGO listing and extracting data
        for ngo in ngo_listings:
            ngo_data = {}

            try:
                ngo_data['ngo_name'] = ngo.find('h3', class_='ngo_name').text.strip()
            except AttributeError:
                ngo_data['ngo_name'] = 'N/A'

            try:
                work_area_tags = ngo.find_all('a', class_='work_area')
                ngo_data['work_areas'] = [area.text.strip() for area in work_area_tags]
            except AttributeError:
                ngo_data['work_areas'] = ['N/A']

            try:
                sub_work_area_tags = ngo.find_all('a', class_='sub_work_area')
                ngo_data['sub_work_areas'] = [area.text.strip() for area in sub_work_area_tags]
            except AttributeError:
                ngo_data['sub_work_areas'] = ['N/A']

            try:
                location_tags = ngo.find_all('a', class_='listing_locations')
                ngo_data['location'] = location_tags[0].text.strip() if location_tags else 'N/A'
            except AttributeError:
                ngo_data['location'] = 'N/A'

            try:
                ngo_data['country'] = location_tags[1].text.strip() if len(location_tags) > 1 else 'N/A'
            except AttributeError:
                ngo_data['country'] = 'N/A'

            try:
                ngo_data['website_url'] = ngo.find('a', itemprop='url')['href'] if ngo.find('a', itemprop='url') else 'N/A'
            except AttributeError:
                ngo_data['website_url'] = 'N/A'

            try:
                ngo_data['facebook_url'] = ngo.find_all('a', itemprop='url')[1]['href'] if len(ngo.find_all('a', itemprop='url')) > 1 else 'N/A'
            except AttributeError:
                ngo_data['facebook_url'] = 'N/A'

            # Add the ngo data to the list
            ngo_data_list.append(ngo_data)

            # Printing the extracted data
            print(f"NGO Name: {ngo_data['ngo_name']}")
            print(f"Work Areas: {', '.join(ngo_data['work_areas'])}")
            print(f"Sub Work Areas: {', '.join(ngo_data['sub_work_areas'])}")
            print(f"Location: {ngo_data['location']}")
            print(f"Country: {ngo_data['country']}")
            print(f"Website: {ngo_data['website_url']}")
            print(f"Facebook: {ngo_data['facebook_url']}")
            print("=" * 50)

    return ngo_data_list

# Call the scraping function to get all the data
ngo_data_list = scrape_ngo_data()
```

```python
def store_data_in_firestore(ngo_data_list):
    # Loop through each NGO data and save to Firebase Firestore
    for ngo_data in ngo_data_list:
        # Create a reference to Firestore collection and document (use NGO name as the document ID)
        ngo_ref = db.collection('ngos').document(ngo_data['ngo_name'])  # Collection: ngos, Document: NGO name
        
        # Save data in Firestore
        ngo_ref.set(ngo_data)

        print(f"Data for {ngo_data['ngo_name']} saved in Firestore!")

# Call the function to store data in Firebase
store_data_in_firestore(ngo_data_list)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_ngo_data():
    ngo_data_list = []
    work_area_list = []
    location_list = []

    # Loop through pages 1 to 8
    for page in range(1, 9):  # Pages 1 to 8
        print(f"Scraping page {page}...")

        # Making a GET request for the current page
        url = f'https://ngobase.org/st/MY.JO/johor-ngos-charities?page={page}'
        r = requests.get(url)

        # Check status code for response received
        if r.status_code != 200:
            print(f"Request failed with status code {r.status_code}")
            continue  # Skip to the next page if the request fails

        # Parsing the HTML
        soup = BeautifulSoup(r.content, 'html.parser')

        # Extracting all NGO entries
        ngo_listings = soup.find_all('div', class_='my_border ngo_listing_div')

        # Looping through each NGO listing and extracting data
        for ngo in ngo_listings:
            ngo_data = {}

            try:
                ngo_data['ngo_name'] = ngo.find('h3', class_='ngo_name').text.strip()
            except AttributeError:
                ngo_data['ngo_name'] = 'N/A'

            try:
                work_area_tags = ngo.find_all('a', class_='work_area')
                ngo_data['work_areas'] = [area.text.strip() for area in work_area_tags]
                work_area_list.extend(ngo_data['work_areas'])  # Collect work areas
            except AttributeError:
                ngo_data['work_areas'] = ['N/A']

            try:
                sub_work_area_tags = ngo.find_all('a', class_='sub_work_area')
                ngo_data['sub_work_areas'] = [area.text.strip() for area in sub_work_area_tags]
            except AttributeError:
                ngo_data['sub_work_areas'] = ['N/A']

            try:
                location_tags = ngo.find_all('a', class_='listing_locations')
                ngo_data['location'] = location_tags[0].text.strip() if location_tags else 'N/A'
                location_list.append(ngo_data['location'])  # Collect locations
            except AttributeError:
                ngo_data['location'] = 'N/A'

            try:
                ngo_data['country'] = location_tags[1].text.strip() if len(location_tags) > 1 else 'N/A'
            except AttributeError:
                ngo_data['country'] = 'N/A'

            try:
                ngo_data['website_url'] = ngo.find('a', itemprop='url')['href'] if ngo.find('a', itemprop='url') else 'N/A'
            except AttributeError:
                ngo_data['website_url'] = 'N/A'

            try:
                ngo_data['facebook_url'] = ngo.find_all('a', itemprop='url')[1]['href'] if len(ngo.find_all('a', itemprop='url')) > 1 else 'N/A'
            except AttributeError:
                ngo_data['facebook_url'] = 'N/A'

            # Add the ngo data to the list
            ngo_data_list.append(ngo_data)

            # Printing the extracted data
            print(f"NGO Name: {ngo_data['ngo_name']}")
            print(f"Work Areas: {', '.join(ngo_data['work_areas'])}")
            print(f"Sub Work Areas: {', '.join(ngo_data['sub_work_areas'])}")
            print(f"Location: {ngo_data['location']}")
            print(f"Country: {ngo_data['country']}")
            print(f"Website: {ngo_data['website_url']}")
            print(f"Facebook: {ngo_data['facebook_url']}")
            print("=" * 50)

    # Remove duplicates by converting lists to sets and back to lists
    work_area_list = list(set(work_area_list))
    location_list = list(set(location_list))

    # Return the complete data and the lists
    return ngo_data_list, work_area_list, location_list

# Call the scraping function to get all the data
ngo_data_list, work_area_list, location_list = scrape_ngo_data()

# Optionally, print the collected work areas and locations without duplicates
print("Unique Work Area List:", work_area_list)
print("Unique Location List:", location_list)
```
